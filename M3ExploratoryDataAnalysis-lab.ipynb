## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here

df['ConvertedComp'].plot(kind='box', figsize=(14,8))
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)

IQR = Q3 - Q1
IQR
```

Find out the upper and lower bounds.


```python
# your code goes here
UpperBound = Q3 + 1.5 * IQR
LowerBound = Q1 - 1.5 * IQR

print('Upper Bound: ', UpperBound)
print('Lower Bound: ', LowerBound)
```

```python
df['ConvertedComp'].describe()
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
df3 = df[['ConvertedComp']] > UpperBound 
print (df3.value_counts())
print (" ")

df4 = df[['ConvertedComp']] < LowerBound 

print (df4.value_counts())

```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here

new_df = df[(df['ConvertedComp'] < UpperBound) | (df['ConvertedComp'] < LowerBound)]
new_df.head()

new_df['ConvertedComp'].mean()
```
